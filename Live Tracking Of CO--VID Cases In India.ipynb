```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df.head()
```

```python
df.drop('sno',axis=1,inplace=True)
```

```python
df
```

```python
df.drop(36,inplace=True)
```

```python
df
```

```python
state = np.arange(len(df['state_name']))
state
```

# Total Positive Cases Of Each State

```python
plt.figure(figsize=(12,10))
plt.barh(state,df['positive'],align='center',color='green')
plt.yticks(state,df['state_name'])
plt.xlabel('Positive Number Of Cases')
plt.title('Corona Virus Cases')
```

 # Total Active Cases Of Each State

```python
plt.figure(figsize=(12,10))
plt.barh(state,df['active'],align='center',color=(1,0,1))
plt.yticks(state,df['state_name'])
plt.xlabel('Active Number Of Cases')
plt.title('Corona Virus Cases')
```

# Total Death Cases Of Each State


```python
plt.figure(figsize=(12,10))
plt.barh(state,df['death'],align='center',color='green')
plt.yticks(state,df['state_name'])
plt.xlabel('Number Of Death Cases')
plt.title('Corona Virus Cases')
```

# Total Number Of Cured Cases Of Each State

```python
plt.figure(figsize=(12,10))
plt.barh(state,df['cured'],align='center',color=(0,0.8,0.8))
plt.yticks(state,df['state_name'])
plt.xlabel('Number Of Cured Cases')
plt.title('Corona Virus Cases')
```
