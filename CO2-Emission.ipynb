```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_excel('gk.xlsx',skiprows = 3,header = 0,index_col=0)
df.fillna(method='ffill')
df.head()
```

```python
z = np.arange(2000,2017)
z
```

```python
df = df.drop([2017,2018,2019,'Country Code', 'Indicator Name', 'Indicator Code'], axis = 1)
df.head()
```

```python
df = df[z]
df
```

```python
df2 = df.T
df2
```

```python
q = df.columns.tolist()
q
```

```python
w = df.loc['India'].tolist()
v = df.loc['United States'].tolist()
u = df.loc['Japan'].tolist()
s = df.loc['Canada'].tolist()
t = df.loc['Switzerland'].tolist()
t
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,2,1.3])
axes.plot(q,w,'-s',label = "INDIA")
axes.plot(q,v,'-o',label = 'United States')
axes.plot(q,u,'-*',label = 'Japan')
axes.plot(q,s,'-x',label = 'Canada')
axes.plot(q,t,'-v',label = 'Switzerland')
axes.set_xlabel('  Years')
axes.set_ylabel('CO2 Emission (metric tons per capita)')
axes.set_title('Comparison of co2 emissions between 5 countries(2000-2016)')
plt.legend()
```
